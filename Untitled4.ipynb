```python
def trapezodial_rule(f,a,b,n):
  h = (b-a)/n
  integration_sum = f(a)+f(b)
  for i in range(1,n):
    k = a+i*h
    integration_sum += 2*f(k)
  return (h/2)*integration_sum
my_function = lambda x:x**2
result = trapezodial_rule(my_function,0,3,100)
print(f"Approximated integral: {result}")
```
