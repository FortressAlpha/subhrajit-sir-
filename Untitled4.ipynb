```python
import numpy as np
import math
def forward_difF(x,y):
  n = len(y)
  table = np.zeros((n,n))
  table[:,0] = y
  for j in range(1,n):
    for i in range(n-j):
      table[i][j] = table[i+1][j-1]-table[i][j-1]
  return table
```

```python
def newton_forward_interpolation(x,y,x_val):
  n_points = len(x) # Store the length in a new variable
  h = x[1] - x[0] # Use the original x array
  table = forward_difF(x,y) # Pass the original x array
  p = (x_val - x[0])/h;
  result = table[0][0]
  p_term = 1
  for i in range(1,n_points): # Use n_points here
    p_term *= (p-(i-1))
    result += (p_term * table[0][i])/math.factorial(i)
  return result,table
```

```python
x_data=np.array([10,20,30,40,50])
y_data = np.array([1.1,2.0,4.4,7.9,12.6])
x_interpolate = 25
result,diff_table = newton_forward_interpolation(x_data,y_data,x_interpolate)
print(diff_table)
print(f"{x_interpolate}{result: .4f}")
```
