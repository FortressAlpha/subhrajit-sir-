```python
import numpy as np
def simpson_one_third(f,a,b,n):
  if n%2 != 0:
    raise ValueError("An even number required")
  h = (b-a)/n
  x = np.linspace(a,b,n+1)
  y = f(x)
  odd_sum = np.sum(y[1:-1:2])
  even_sum = np.sum(y[2:-1:2])
  integral = (h/3)*(y[0]+4*odd_sum+2*even_sum+y[-1])
  return integral
```

```python
def my_function(x):
  return (x**2)*np.exp(x)
lower_limit = 0.0
upper_limit = 2.0
intervals = 10
result = simpson_one_third(my_function,lower_limit,upper_limit,intervals)
print(f"Approximated value: {result:.6f}")
```
