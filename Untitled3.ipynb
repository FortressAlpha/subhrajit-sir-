```python
import numpy as np
import math
def backward_diff_table(y):
  n = len(y)
  table=np.zeros((n,n))
  table[:,0] = y
  for j in range(1,n):
    for i in range(n-1,j-1,-1):
      table[i][j] = table[i][j-1] - table[i-1][j-1]
  return table
```

```python
def newton_backward_interpolation(x,y,x_val):
  n = len(x)
  h = x[1] - x[0]
  table = backward_diff_table(y)
  p = (x_val - x[-1])/h
  result = table[n-1,0]
  p_term = 1
  for i in range(1,n):
    p_term *= (p+(i-1))
    result += (p_term*table[n-1,i])/math.factorial(i)
  return result,table
```

```python
x_data = np.array([10,20,30,40,50],dtype = float)
y_data = np.array([1.1,2.0,4.4,7.9,12.6],dtype = float)
x_interpolate = 45
result,diff_table = newton_backward_interpolation(x_data,y_data,x_interpolate)
print(diff_table)
print(f"{x_interpolate}: {result:.4f}")
```
